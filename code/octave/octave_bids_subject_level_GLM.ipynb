Unzip and smooth the data

```octave
data_dir = '../inputs/NARPS/';
fmriprep_dir = '../inputs/NARPS/derivatives/fmriprep/';

space_label = 'MNI152NLin2009cAsym';
task_label = 'MGT';

subject = '001';

FWHM = 6;

% let's ouput the data into a different derivative folder to keep the original data clean
output_dir = '../inputs/NARPS/derivatives/spm12/';
spm_mkdir(output_dir, ['sub-' subject], {'anat', 'func'});
subject_ouput_dir = fullfile(output_dir, ['sub-' subject], 'func');                   
```

```octave
% select the preprocessed data
subject_input_dir = fullfile(fmriprep_dir, ['sub-' subject]);

input_files = spm_select('FPlist', ...
                         fullfile(subject_input_dir, 'func'), ...
                         [space_label '.*_preproc.nii.gz'])
                         

%% Step 1: unzip the file
matlabbatch{1}.cfg_basicio.file_dir.file_ops.cfg_gunzip_files.files = cellstr(input_files);
matlabbatch{1}.cfg_basicio.file_dir.file_ops.cfg_gunzip_files.outdir = {subject_ouput_dir};
matlabbatch{1}.cfg_basicio.file_dir.file_ops.cfg_gunzip_files.keep = true;

% Step 2: smooth the files we have unzipped
% OK this next lines look scary but what it does is basically:
%   - grabs the ouput of step 1
%   - makes sure it is used as input of step 2
matlabbatch{2}.spm.spatial.smooth.data(1) = cfg_dep('Gunzip Files: Gunzipped Files', ...
                                            substruct('.','val', '{}',{1}, ...
                                                      '.','val', '{}',{1}, ...
                                                      '.','val', '{}',{1}, ...
                                                      '.','val', '{}',{1}), ...
                                                      substruct('()',{':'}));
matlabbatch{2}.spm.spatial.smooth.fwhm = [FWHM FWHM FWHM]; 
```

## Prepare the model

### Get the events files

The `events.tsv` files that described what happened when during the fMRI experiment are in the original dataset and not in the fMRIprep derivatives.

Let's list them and load their content.

```octave
subject_input_dir = fullfile(data_dir, ['sub-' subject])

events_files = spm_select('FPlist', ...
                         fullfile(subject_input_dir, 'func'), ...
                         '.*_events.tsv')

for iRun = 1:size(events_files, 1)
    events(iRun) = spm_load(events_files(iRun,:));
end

events
```
